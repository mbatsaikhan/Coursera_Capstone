```python
#use this function to open the downloaded data set
df = pd.read_csv("FuelConsumption.csv")
#use this function to take a look at the dataset
df.head()
```

```python
#summarize the data
df.describe()
```

```python
#create a new data set named cdf by picking out certain columns from original dataset df
cdf = df[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB','CO2EMISSIONS']]
cdf.head(9)
```

```python
#plotting each of the columns from cdf, histogram plotting
viz = cdf[['CYLINDERS','ENGINESIZE','CO2EMISSIONS','FUELCONSUMPTION_COMB']]
#histograms show how common a sinlge value is recurring
#the highest bars are the most frequent values
#the shortest bars are the least frequent variables in a list
viz.hist()
plt.show()
```

```python
#now we plot the emission feature of cdf in relation to fuel consumption
#this is showing a linear relationship
plt.scatter(cdf.FUELCONSUMPTION_COMB, cdf.CO2EMISSIONS,  color='blue')
plt.xlabel("FUELCONSUMPTION_COMB")
plt.ylabel("Emission")
plt.show()
```

```python
#now we plot the emission feature of cdf in relation to Engine Size
#this is showing a linear relationship
plt.scatter(cdf.ENGINESIZE, cdf.CO2EMISSIONS,  color='blue')
plt.xlabel("Engine size")
plt.ylabel("Emission")
plt.show()
```

```python
#now we plot the emission feature of cdf in relation to Cylinders
#this is showing a linear relationship
plt.scatter(cdf.CYLINDERS, cdf.CO2EMISSIONS, color='blue')
plt.xlabel("Cylinders")
plt.ylabel("Emission")
plt.show()
```

```python
#Creating a dataset named train and another dataset named test
#Train/Test Split will split the dataset into training and testing sets
#You will use the training partition to train your model and use the test set to test your model.
#We will be using 80/20 to split the data. 80% for training and 20% for testing.
#We create a mask to select random rows using np.random.rand()
msk = np.random.rand(len(df)) < 0.8
train = cdf[msk]
test = cdf[~msk]
```

```python
#
#now we plot the emission feature of dataset train in relation to Engine Size
#this is showing a linear relationship for the trained dataset
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS,  color='blue')
#assigning name to x axis of plot
plt.xlabel("Engine size")
#assigning name to y axis of plot
plt.ylabel("Emission")
plt.show()
```

```python
#Modeling
#using sklearn package to model our data above
from sklearn import linear_model
regr = linear_model.LinearRegression()
train_x = np.asanyarray(train[['ENGINESIZE']])
train_y = np.asanyarray(train[['CO2EMISSIONS']])
regr.fit (train_x, train_y)
#The coefficients
#Coefficient and Intercept in the simple linear regression, are the parameters of the fit line
#knowing that the parameters are hte intercept and slope of the line, sklearn can estimate them
print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
```

```python
#Plot Outputs
#now we will plot the new fit line over the data. 
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS,  color='blue')
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], '-r')
plt.xlabel("Engine size")
plt.ylabel("Emission")
```

```python
#Evaluation
#we will compare the actual values and predicted values to calculate the accuracy of our regression.
#we will use MSE Mean Squared Error to calculate the accuracy of our model based on the test dataset.
from sklearn.metrics import r2_score

test_x = np.asanyarray(test[['ENGINESIZE']])
test_y = np.asanyarray(test[['CO2EMISSIONS']])
test_y_hat = regr.predict(test_x)

#show the Mean Absolute Error 
print("Mean absolute error: %.2f" % np.mean(np.absolute(test_y_hat - test_y)))
#show our Mean Square Error
print("Residual sum of squares (MSE): %.2f" % np.mean((test_y_hat - test_y) ** 2))
#show our 
print("R2-score: %.2f" % r2_score(test_y_hat , test_y) )
```
