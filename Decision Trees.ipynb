```python
#importing necessary libraries
import numpy as np 
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
!wget -O drug200.csv https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/ML0101ENv3/labs/drug200.csv
```

```python
my_data = pd.read_csv("drug200.csv", delimiter=",")
my_data[0:5]
```

```python
#removing the column containing target name since it doesn't contain numeric value
X = my_data[['Age', 'Sex', 'BP', 'Cholesterol', 'Na_to_K']].values
X[0:5]
```

```python
#because some of the features in the dataset were catagorical such as sex and blood pressures
#we have to convert then into numerical features
#pandas.get_dummies() converts categorical variables into dummy/ indicator variables
from sklearn import preprocessing
le_sex = preprocessing.LabelEncoder()
le_sex.fit(['F','M'])
X[:,1] = le_sex.transform(X[:,1]) 


le_BP = preprocessing.LabelEncoder()
le_BP.fit([ 'LOW', 'NORMAL', 'HIGH'])
X[:,2] = le_BP.transform(X[:,2])


le_Chol = preprocessing.LabelEncoder()
le_Chol.fit([ 'NORMAL', 'HIGH'])
X[:,3] = le_Chol.transform(X[:,3]) 

X[0:5]
```

```python
#now we can fill the target variable
y = my_data["Drug"]
y[0:5]
```

```python
#we'll use the train/test split on our decision tree
#importing the train_test_split from sklearn.cross_validation
from sklearn.model_selection import train_test_split
```

```python
#train_test_split will return 4 different parameters: x_trainset,x_testset,y_trainset,y_testset
#the train_test_split will need the parameters: X,Y,test_size=0.3,and random_state=3
X_trainset, X_testset, y_trainset, y_testset = train_test_split(X, y, test_size=0.3, random_state=3)
```

```python
#modeling, we will first creat an instance of the decision tree called drugTree
#inside the classifier, specify criterion="entropy" so we can see the information dain of nodes
drugTree = DecisionTreeClassifier(criterion="entropy", max_depth = 4)
drugTree # it shows the default parameters
```

```python
# now we fit the data with training feature matrix X_trainset and training response vector y_trainset
drugTree.fit(X_trainset,y_trainset)
```

```python
#Prediction, lets make some predictiopns in the testing dataset and store it into a variable called predTree
predTree = drugTree.predict(X_testset)
```

```python
print (predTree [0:5])
print (y_testset [0:5])
```

```python
#Evaluation, import metrics from sklearn and check the accuracy of our model
from sklearn import metrics
import matplotlib.pyplot as plt
print("DecisionTrees's Accuracy: ", metrics.accuracy_score(y_testset, predTree))
```
